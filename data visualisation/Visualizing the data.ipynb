```python
def f(x,y):
    return np.sin(np.sqrt(x**2+y**2))

x=np.linspace(-6,6,50)
y=np.linspace(-6,6,50)
x,y=np.meshgrid(x,y)
z=f(x,y)
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.contour3D(x,y,z,cmap='binary')
```
