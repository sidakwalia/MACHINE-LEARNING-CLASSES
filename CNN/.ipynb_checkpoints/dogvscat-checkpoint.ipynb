```python
import numpy as np
import pandas as pd
from keras.preprocessing.image import ImageDataGenerator,load_img #For data augmentation
from keras.utils import to_categorical #Convert to 1000, 0100 and so on to reduce size
from sklearn.model_selection import train_test_split #To split dataset
import matplotlib.pyplot as plt
import random 
import os
print(os.listdir("../input/dogs-vs-cats/test/test"))
```

```python
print(len(os.listdir('../input/dogs-vs-cats/train/train')))
print(len(os.listdir('../input/dogs-vs-cats/test/test')))
```

```python
print(os.listdir('../input/dogs-vs-cats/train/train'))
```

```python
filenames=os.listdir("../input/dogs-vs-cats/train/train/") #Open the folder
categories=[]
for filename in filenames:
    category=filename.split('.')[0] 
    if category=='dog': #If it is a dog, write 1
        categories.append(1)
    else: #If it is a cat, write 0
        categories.append(0)
#Create a dataframe to store the filename and the category        
df=pd.DataFrame({ 
    'filename':filenames,
    'category':categories
})
```

```python
df.tail()
```

```python
df['category'].value_counts().plot.bar()
```

```python
sample=random.choice(filenames)
image=load_img("train/" +sample)
plt.imshow(image) #show a randow image
```

```python
from keras.models import Sequential #For creating a sequential model. ie-layer by layer model
from keras.layers import Conv2D,MaxPooling2D,Dropout,Flatten,Dense,Activation #Required to create a CNN 
```

```python
image_width=128
image_height=128
image_size=(image_width,image_height) #Resize the image
image_channels=3 #Three colours in the image
```

```python
model=Sequential() #Create a CNN
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(image_width,image_height,image_channels)))
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.25))
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.25))
model.add(Conv2D(128,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2,activation='softmax'))              #model.add(Dense(2,activation='softmax'))  

model.compile(loss='categorical_crossentropy',optimizer='Adam',metrics=['accuracy']) #Use Adam as the optimizer and show accuracy as well

model.summary()
```

```python
from keras.callbacks import EarlyStopping,ReduceLROnPlateau
```

```python
earlystop=EarlyStopping(patience=10) #Stop the model after a specified time frame
```

```python
learning_rate_reduce=ReduceLROnPlateau(monitor='val_acc',min_lr=0.0001) #reduce learning rate after sometime
```

```python
callbacks=[earlystop,learning_rate_reduce]
```

```python
df['category']=df['category'].replace({0:'cat',1:'dog'}) #Replace 0 and 1 with cats and dogs
```

```python
train_df,validate_df=train_test_split(df,test_size=0.20) #Split training and testing data by 8:2
train_df=train_df.reset_index(drop=True) #Doubt
validate_df=validate_df.reset_index(drop=True) #Doubt
```

```python
train_df['category'].value_counts().plot.bar()
```

```python
validate_df['category'].value_counts().plot.bar()
```

```python
total_train=train_df.shape[0]
total_validate=validate_df.shape[0]

print(total_train,total_validate) #Print out the number of images(validation and training)
```

```python
train_datagen=ImageDataGenerator(
rotation_range=15,
rescale=1/255,
zoom_range=0.1,
horizontal_flip=True,width_shift_range=0.1,height_shift_range=0.1) #Apply image augmentation 

train_generator=train_datagen.flow_from_dataframe(train_df,"../input/dogs-vs-cats/train/train/",x_col='filename',y_col='category',target_size=image_size,class_mode='categorical',batch_size=64)#Prepare training data and change batch size, target size etc.
```

```python
validation_datagen=ImageDataGenerator(
rotation_range=15,
rescale=1/255,
zoom_range=0.1,
horizontal_flip=True,width_shift_range=0.1,height_shift_range=0.1) #Apply image augmentation 

valid_generator=validation_datagen.flow_from_dataframe(validate_df,"../input/dogs-vs-cats/train/train/",x_col='filename',y_col='category',target_size=image_size,class_mode='categorical',batch_size=64)#Prepare validataion data and change batch size, target size etc.
```

```python
example_df=train_df.sample(n=1).reset_index(drop=True)
example_generator=train_datagen.flow_from_dataframe(
example_df,"../input/dogs-vs-cats/train/train/",x_col='filename',y_col='category',target_size=image_size,class_mode='categorical'


)
```

```python
batch_size=250 #set batch size
history=model.fit_generator(train_generator,epochs=20,validation_data=valid_generator,validation_steps=total_validate/batch_size,steps_per_epoch=total_train/batch_size,callbacks=callbacks)#Train the model
```

```python
plt.plot(history.history['val_loss'],color='r')
plt.plot(history.history['accuracy'],color='black')
plt.plot(history.history['loss'],color='black')
plt.plot(history.history['val_accuracy'],color='yellow')
plt.show() #Plot the model loss, accuracy, validation loss and validation accuracy
```
