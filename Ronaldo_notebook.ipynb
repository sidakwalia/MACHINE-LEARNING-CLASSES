```python
#exploration of data and its anaylsis
data.drop(['Unnamed: 0','remaining_min.1','power_of_shot.1','knockout_match.1','remaining_sec.1'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.date_of_game=pd.to_datetime(data.date_of_game)
data['game_season']=data['game_season'].astype('object')
data['game_season']
```

```python
#label encoding
l_unique=data['game_season'].unique()
l_unique
```

```python
v_unique=np.arange(len(l_unique))
v_unique
```

```python
data['game_season'].replace(to_replace=l_unique,value=v_unique,inplace=True)
```

```python
data.head()
```

```python
#filling the nan values with mean
data['power_of_shot'].fillna(value=data['power_of_shot'].mean(),inplace=True)
data.isnull().sum()
```

```python
#filling nan values with mode
mode_data=data.type_of_combined_shot.value_counts().keys()[0]
print('mode is',mode_data)
data.type_of_combined_shot.fillna(value=mode_data,inplace=True)
data.isnull().sum()
```

```python
data.remaining_sec.fillna(value=data.remaining_sec.median(),inplace=True)
data.isnull().sum()
```

```python
data['location_x'].fillna(value=0,inplace=True)
data['location_y'].fillna(value=0,inplace=True)
data.isnull().sum()
```

```python
print('NULL VALUES IN column home/away=',data['home/away'].isnull().sum())
col=['home/away','lat/lng','team_name','remaining_min']
data.loc[:,col]=data.loc[:,col].ffill()
data.isnull().sum()
```
